```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
from selenium.webdriver.common.by import By

# ì£¼ì˜! runìœ¼ë¡œë§Œ ì‹¤í–‰í•  ê²ƒ
```

```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/" + str(word)
    return url
```

```python
def select_first(driver):
    first = driver.find_elements(By.CSS_SELECTOR, "div._aagw")[0]
    first.click()
    time.sleep(3)
```

```python
import re
from bs4 import BeautifulSoup

def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    # ë³¸ë¬¸ ë‚´ìš©
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''
    # í•´ì‹œíƒœê·¸
    tags = re.findall(r'#[^\s#,\\]+', content)
    
    # ì‘ì„±ì¼ì
    date = soup.select('time._aaqe')[0]['datetime'][:10]
    
    # ìœ„ì¹˜
    try:
        place = soup.select('div._aaqm')[0].text
    except:
        place = ''

    data = [date, place, tags]
    return data
```

```python
def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR, "div._aaqg._aaqh") # 2022/01/11 ìˆ˜ì •
    right.click()
    time.sleep(3)
```

```python
# í¬ë¡¤ë§ ì‹œì‘
"""
driver.get(url)ì„ í†µí•´ ê²€ìƒ‰ í˜ì´ì§€ ì ‘ì†í•˜ê³ ,
target ë³€ìˆ˜ì— í¬ë¡¤ë§í•  ê²Œì‹œê¸€ì˜ ìˆ˜ë¥¼ ë°”ì¸ë”©
"""

# í¬ë¡¬ ë¸Œë¼ìš°ì € ì—´ê¸°
driver = webdriver.Chrome()

driver.get('https://www.instagram.com')

time.sleep(5)

# ì¸ìŠ¤íƒ€ê·¸ë¨ ë¡œê·¸ì¸ì„ ìœ„í•œ ê³„ì • ì •ë³´
email = #ì¸ìŠ¤íƒ€ê·¸ë¨ ì´ë©”ì¼ ìˆ˜ë™ ì…ë ¥
input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
input_id.clear()
input_id.send_keys(email)

password = #ì¸ìŠ¤íƒ€ê·¸ë¨ ì´ë©”ì¼ ìˆ˜ë™ ì…ë ¥
input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
input_pw.clear()
input_pw.send_keys(password)
input_pw.submit()

time.sleep(5)

```

```python
# ê²Œì‹œë¬¼ì„ ì¡°íšŒí•  ê²€ìƒ‰ í‚¤ì›Œë“œ ì…ë ¥ ìš”ì²­
word = input("ê²€ìƒ‰ì–´ë¥¼ ì…ë ¥í•˜ì„¸ìš” : ")
word = str(word)
url = insta_searching(word)

# ê²€ìƒ‰ ê²°ê³¼ í˜ì´ì§€ ì—´ê¸°
driver.get(url)
time.sleep(10) # ì½”ë“œ ìˆ˜í–‰ í™˜ê²½ì— ë”°ë¼ í˜ì´ì§€ê°€ ë¡œë“œë˜ëŠ” ë° ì‹œê°„ì´ ë” ê±¸ë¦´ ìˆ˜ ìˆì–´ 8ì´ˆë¡œ ë³€ê²½(2022/01/11)

# ì²« ë²ˆì§¸ ê²Œì‹œë¬¼ í´ë¦­
select_first(driver)

# ë³¸ê²©ì ìœ¼ë¡œ ë°ì´í„° ìˆ˜ì§‘ ì‹œì‘
results = []
## ìˆ˜ì§‘í•  ê²Œì‹œë¬¼ì˜ ìˆ˜
target = 10
for i in range(target):
    print(i)

    try:
        data = get_content(driver)
        results.append(data)
        move_next(driver)
    except:
        time.sleep(2)
        move_next(driver)
    time.sleep(5)

print(results)
```
